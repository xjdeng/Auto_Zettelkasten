```python
#see: https://stackoverflow.com/questions/64743583/which-10-words-has-the-highest-tf-idf-value-in-each-document-total
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from collections import defaultdict
import numpy as np
import re
import heapq
from sklearn.feature_extraction import text
import joblib
```

```python
N = 10
```

```python
def second_largest(arr):
    res = heapq.nlargest(2, arr)
    return res[1]
```

```python
my_stop_words = text.ENGLISH_STOP_WORDS.union(["_","__","___"])

X,y = fetch_20newsgroups(return_X_y = True)
#X = X[0:1000]
tfidf = TfidfVectorizer(stop_words=my_stop_words)
X_tfidf = tfidf.fit_transform(X)#.toarray()
docs,words = X_tfidf.shape
doclookup = [[] for _ in range(docs)]
wordlookup = [[] for _ in range(words)]
#collookup = defaultdict(list)
row,col = X_tfidf.nonzero()
for r,c in zip(row,col):
    doclookup[r].append(c)
    wordlookup[c].append(r)
vocab = tfidf.vocabulary_
reverse_vocab = {v:k for k,v in vocab.items()}

feature_names = tfidf.get_feature_names()

best_tfidf = []

for i,doc in enumerate(doclookup):
    myset = set()
    myrow = list(X_tfidf[i,doclookup[i]].toarray()[0])
    reverse = {item:idx for item,idx in zip(myrow, doclookup[i])}
    descending = sorted(myrow, reverse=True)
    for item_idx in descending:
        if len(wordlookup[reverse[item_idx]]) > 1:
            myset.add(reverse_vocab[reverse[item_idx]])
            if len(myset) == N:
                break
    best_tfidf.append(myset)
```
